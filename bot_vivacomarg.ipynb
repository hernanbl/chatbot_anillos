3: Cargar y Procesar el PDF

```python
# Ruta del PDF (ajusta según donde esté tu archivo)
pdf_path = "/content/sample_data/juego-tronos.pdf"  # Cambia "tu_pdf.pdf" por el nombre de tu archivo

# Verificar si el PDF existe
if not os.path.exists(pdf_path):
    raise FileNotFoundError(f"No se encontró el archivo PDF en: {pdf_path}")

# Cargar y procesar el PDF
loader = PyPDFLoader(pdf_path)
pages = loader.load_and_split()

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200
)
docs = text_splitter.split_documents(pages)

# Crear vector store (base de datos de embeddings)
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
vector_store = FAISS.from_documents(docs, embeddings)

print("PDF cargado y procesado correctamente.")
```

**4: Función para Hacer Preguntas**

```python
def ask_with_context(query):
    """
    Responde preguntas basadas en el contenido del PDF.
    """
    # Buscar los fragmentos relevantes del PDF
    relevant_docs = vector_store.similarity_search(query, k=10)
    context = "\n".join([doc.page_content for doc in relevant_docs])

    # Crear prompt más flexible
    prompt = f"""
    Eres un asistente que responde principalmente con información del siguiente contexto.
    Si la pregunta no puede responderse directamente, intenta inferir una respuesta basada en el contexto.
    Si no hay información relevante, di: "No tengo suficiente información para responder eso".

    Contexto:
    {context}

    Pregunta: {query}

    ### Respuesta:""" # Added ### to clearly delimit the answer section

    # Generar respuesta con Gemini
    model = genai.GenerativeModel('gemini-pro')
    response = model.generate_content(prompt)

    # Handle cases where 'text' attribute is not available
    try:
        answer = response.text
    except ValueError:
        # If 'text' is not available, try accessing it from candidates
        if response.candidates and response.candidates[0].content:
            answer = response.candidates[0].content.text
        else:
            answer = "Gemini was unable to provide an answer." # Provide a default response

    return answer
```

5: Hacer Preguntas, última celda

```python
# Ejemplo de uso
pregunta = "Quién era Theon Greyjoy y qué papel desempeñaba?"  # <-- Hacer una pregunta acá
respuesta = ask_with_context(pregunta)
print("Respuesta:", respuesta)
```
